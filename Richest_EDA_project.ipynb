```python
def value_to_float(x):
    if 'K' in x :
        if len(x) > 1 :
            return float(x.replace('K',''))*1000
    if 'M' in x :
        if len(x) > 1 :
            return float(x.replace('M',''))*1000000
    if 'B' in x :
        if len(x) > 1 :
            return float(x.replace('B',''))*1000000000
data['$ Last Change'] = data['$ Last Change'].apply(value_to_float)            
```
